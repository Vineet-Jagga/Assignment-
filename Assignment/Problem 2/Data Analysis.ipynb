```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\Vineet\\Downloads\\AB_NYC_2019.csv")
```

```python
df.sample(10)
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df['Price/night($)'])
plt.show()
```

```python
print("Mean value of Price/night($)", df['Price/night($)'].mean())
print("Std value of Price/night($)", df['Price/night($)'].std())
print("Min value of Price/night($)", df['Price/night($)'].min())
print("Max value of Price/night($)", df['Price/night($)'].max())
print("variance of Price/night($)", df['Price/night($)'].var())
```

```python
# finding boundary values
print("Highest price", df['Price/night($)'].mean() + 3*df['Price/night($)'].std())
print("lowest price", df['Price/night($)'].mean() -  3*df['Price/night($)'].std())
```

```python
# finding the outliers
df[(df['Price/night($)']>533.51) | (df['Price/night($)']<-229.01)]
```

```python
# removing the outliers
new_df = df[(df['Price/night($)'] < 533.51 )  &  (df['Price/night($)'] > -229.01)]
new_df
```
